```python
#MATplotib, seaBorn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
bulan = ['Januari', 'Febuari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
minggu_1 = np.random.randint(200000,700001,12)
minggu_2 = np.random.randint(200000,700001,12)
minggu_3 = np.random.randint(200000,700001,12)
minggu_4 = np.random.randint(200000,700001,12)

data = {
    'Bulan' : bulan,
    'minggu_1': minggu_1,
    'minggu_2': minggu_2,
    'minggu_3': minggu_3,
    'minggu_4' : minggu_4
}
df = pd.DataFrame(data)
df['Total_pengeluaran'] = df[['minggu_1','minggu_2','minggu_3','minggu_4']].sum(axis=1)
df
```

```python
#Diagram Bar
#Buat canvas
plt.figure(figsize=(12,6)) #Widht, height -> inch
plt.bar(df['Bulan'],df['Total_pengeluaran'],color='skyblue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.title("Total Pengeluaran Mahasiswa Per Bulan")
plt.ylabel("Rupiah")
plt.show()
```

```python
#Tren pengeluaran per minggu dalam satu tahun

plt.figure(figsize=(10,5))

for minggu in ['minggu_1','minggu_2','minggu_3','minggu_4']:
    plt.plot(df['Bulan'],df[minggu],marker='o',label=minggu)

plt.title("Tren Pengelolahan Mahasiswa per Minggu")
plt.ylabel("Rupiah")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Diagram Pie ( Proporsi Pengeluaran Per Bulan)
plt.figure(figsize=(6,6))
plt.pie(df['Total_pengeluaran'],labels=df['Bulan'],startangle=90, autopct='%1.1f%%',colors=plt.cm.Pastel1.colors)
plt.axis('equal')
plt.title("Proporsi Pengeluaran Mahasiswa Per Bulan")
plt.show()
```

```python
#Model Ai ->scikit , sklearn
from sklearn.linear_model import LinearRegression
df['bulan'] = np.arange(1,13)

x = df[['bulan']]
y = df['Total_pengeluaran']

model = LinearRegression()
model.fit(x,y)

prediksi_13 = model.predict([[13]])[0]
print(f"Prediksi Pengeluaran Mahasiswa bulan Januari: {prediksi_13:,.0f}")
plt.figure(figsize=(10,10))
plt.boxplot(df['Total_pengeluaran'])
plt.show()
```
